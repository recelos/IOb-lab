# Zadanie 1

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = (18, 10)
image = cv.imread("images/example.jpg")
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)


kernel = np.array([[-1,-1,-1],
                   [-1, 8,-1],
                   [-1,-1,-1]])

filtered = cv.filter2D(image, -1, kernel)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[1].imshow(filtered)
```

# Zadanie 2

```python
colored = image.copy()
colored = colored / 255.0

kernel = np.array([[0.393, 0.769, 0.189],
                   [0.349, 0.689, 0.168],
                   [0.272, 0.534, 0.131]])

colored = cv.transform(colored, kernel)
colored = np.clip(colored, 0, 1)
colored = colored * 255.0
colored = colored.astype(np.uint8)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[1].imshow(colored)
```

# Zadanie 4

```python
imageYCrCb = image.copy()

imageYCrCb = imageYCrCb.astype(np.float32)

kernel = np.array([[ 0.229,  0.587,  0.114],
                   [ 0.500, -0.418, -0.082],
                   [-0.168, -0.331,  0.500]])

imageYCrCb = cv.transform(imageYCrCb, kernel)
        
imageYCrCb[:, :, 1] += .5     
imageYCrCb[:, :, 2] += .5    

imageYCrCb = np.clip(0, 255, imageYCrCb)

fig, ax = plt.subplots(1, 3)
ax[0].imshow(image)
ax[1].imshow(imageYCrCb)
imageRGB = cv.cvtColor(imageYCrCb, cv.COLOR_YCrCb2RGB)
ax[2].imshow(imageRGB)
```
